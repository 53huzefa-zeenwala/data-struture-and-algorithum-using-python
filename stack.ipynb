```python
class Stack_arr:
    def __init__(self):
        self.arr = []
        
    def push(self, data):
        self.arr.append(data)
        
    def pop(self):
        return self.arr.pop()
        
    def peek(self):
        if self.arr == []:
            return "stack is empty"
        
        return self.arr[-1]
    
    def push_at_end(self, data):
        if self.arr == []:
            self.arr.append(data)
            return
            
        last = self.arr.pop()
        self.push_at_end(data)
        
        self.arr.append(last)
        
    def display(self):
        for i in self.arr:
            print(i, end=" ")
    
    def reverse(self):
        if self.arr == []:
            return 
        
        last = self.arr.pop()
        self.reverse()
        
        self.push_at_end(last)
        
            
```
